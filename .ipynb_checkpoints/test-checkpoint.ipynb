```python
from FEM1D import fem1D
from FEM1D import build_solution
from FEM1D import erro_L2
import numpy as np
import matplotlib.pyplot as plt
import exemplos
from IPython.display import display, Math
```

```python
n = 25
quad=[]
cubic=range(n)
problema = exemplos.exemplo5(0.01)

alpha = problema.alpha
beta = problema.beta
gamma = problema.gamma
f = problema.f
a = problema.a
b = problema.b
u0 = problema.ua
uN = problema.ub

X = np.linspace(a,b,n+1+len(quad)+2*len(cubic))

U, Elements = fem1D(n, alpha, beta, gamma, f, a, b, u0, uN, X, quad, cubic)
```

```python
#Gera a solução numérica, a sua derivada e os pontos para o plot
sol_num, sol_diff, z = build_solution(U, Elements)

#Solução e derivada analítica
u = problema.solucao
du = problema.derivada

#Erro da solução e erro da derivada
e0 , de0 = erro_L2(U,u,du,Elements)

fig, (ax1,ax2) = plt.subplots(1,2)
fig.set_figheight(7)
fig.set_figwidth(11)

#Plot da solução

ax1.plot(z, u(z), label='Sol. Exata.', lw=4)
ax1.plot(z, sol_num, label='FEM', lw=2)
ax1.set_xlabel('X',fontsize=14)
ax1.grid()
ax1.legend()

#Plot da derivada
ax2.plot(z, du(z), label='Derivada Exata', lw=4)
ax2.plot(z, sol_diff, label='FEM', lw=2)
ax2.set_xlabel('X', fontsize=14)
ax2.grid()
ax2.legend()

display(Math(r'||u-U||_0 = {:.7f}'.format(e0)))
display(Math(r'\Bigg|\Bigg|\frac{du}{dx} - \frac{dU}{dx}\Bigg|\Bigg|_1 '+'= {:.7f}'.format(de0)))

fig.savefig('quadratico10.jpg')
```

```python
#NESTA CÉLULA EU FIZ UM LOOP PARA PLOTAR TODOS OS ERROS DE UMA SÓ VEZ

N_l = 30
N_c = 30
N_q = 30

fig, (ax1, ax2) = plt.subplots(2,1)
fig.set_figheight(9)
fig.set_figwidth(9)

ax1.grid(which="both")
ax1.set_xlabel('log(h)')
ax1.set_ylabel('$log ||u - U||_0$')
ax1.set_yscale('log')
ax1.set_xscale('log')
ax1.set_xticks([0.1, 0.3, 0.5, 0.7])
ax1.get_xaxis().get_major_formatter().labelOnlyBase = False

ax2.set_yscale('log')
ax2.set_xscale('log')
ax2.grid(which="both")
ax2.set_xlabel('log(h)')
ax2.set_ylabel('$log ||u - U||_1$')
ax2.set_xticks([0.1, 0.3, 0.5, 0.7])
ax2.get_xaxis().get_major_formatter().labelOnlyBase = False

for ord in [1,2,3]:
    
    if ord==1:
        H = []
        E = []
        dE = []
        for k in range(3,N_l+1):
        
            X = np.linspace(a,b,k+1)
            U, Elements = fem1D(k, alpha, beta, gamma, f, a, b, u0, uN, X, quad=[],cubic=[])
            e0, de0 = erro_L2(U, u, du, Elements) 
            h = Elements[0].h
            H.append(h)
            E.append(e0)
            dE.append(de0)

        ax1.plot(H, E, lw=2, label='linear')
        ax2.plot(H, dE,lw=2, label='linear')

        #print((np.log(E[-1])-np.log(E[-2]))/(np.log(H[-1])-np.log(H[-2])))
        #print((np.log(dE[-1])-np.log(dE[-2]))/(np.log(H[-1])-np.log(H[-2])))

    elif ord==2:
        H = []
        E = []
        dE = []
        for k in range(3,N_q+1):
        
            quad = range(k)
            X = np.linspace(a,b,k+1+len(quad))
            U, Elements = fem1D(k, alpha, beta, gamma, f, a, b, u0, uN, X, quad=range(k),cubic=[])
            e0, de0 = erro_L2(U, u, du, Elements) 
            h = Elements[0].h
            H.append(h)
            E.append(e0)
            dE.append(de0)

        ax1.plot(H, E, lw=2, label='quad')
        ax2.plot(H, dE, lw=2, label='quad')

        #print((np.log(E[-1])-np.log(E[-2]))/(np.log(H[-1])-np.log(H[-2])))
        #print((np.log(dE[-1])-np.log(dE[-2]))/(np.log(H[-1])-np.log(H[-2])))
    
    elif ord==3:
        H = []
        E = []
        dE = []
        for k in range(3,N_c+1):
            
            cubic = range(k)
            X = np.linspace(a,b,k+1+2*len(cubic))
            U, Elements = fem1D(k, alpha, beta, gamma, f, a, b, u0, uN, X, quad=[],cubic=range(k))
            e0, de0 = erro_L2(U, u, du, Elements) 
            h = Elements[0].h
            H.append(h)
            E.append(e0)
            dE.append(de0)

        ax1.plot(H, E, lw=2, label='cúbico')
        ax2.plot(H, dE, lw=2, label='cúbico')

        #print((np.log(E[-1])-np.log(E[-2]))/(np.log(H[-1])-np.log(H[-2])))
        #print((np.log(dE[-1])-np.log(dE[-2]))/(np.log(H[-1])-np.log(H[-2])))
        
ax1.legend()
ax2.legend()
```
